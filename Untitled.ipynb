```python
%matplotlib inline

import libpysal as lp
import numpy as np
import xarray as xr
from esda import Moran_Local
```

```python
da = xr.open_rasterio('HARV_RGB_Ortho.tif')
```

```python
da = da[:, 600:700, 700:800]
da
```

```python
da[0:1].plot()
```

```python
w_rook = lp.weights.Rook.from_xarray(da)
```

```python
w_rook.n
```

```python
w_rook.ids
```

```python
data = da.to_series()[w_rook.ids]
data
```

```python
np.random.seed(12345)
lisa = Moran_Local(data,w_rook)
```

```python
moran_da = lp.weights.toDataArray(lisa.p_sim, w_rook.id_order, da.coords, da.attrs)
```

```python
moran_da.plot()
```
